```python
import pandas as pd # Manejo y análisis de estructuras de datos
import numpy as np # Cálculo numérico y el análisis de datos
import seaborn as sns # Creación gráficos estadísticos
import matplotlib.pyplot as plt # Creación de gráficos en dos dimensiones

# librerías de para implementar modelos de ciencia de datos
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import RocCurveDisplay
```

```python
Datos = pd.read_csv('C:/Users/Natalia G/Desktop/Titanic.csv')
```

```python
Datos.head(40)
```

```python
Datos.describe()
```

```python
print(Datos.isnull().sum())
```

```python
sns.countplot(x='Survived', data=Datos)
plt.title('Distribución de Sobrevivientes')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['Age'])
plt.title('Identificación de Datos Atípicos en Edad')
plt.show()
```

#### 2. Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.

```python
Datos['Age'] = Datos['Age'].fillna(Datos['Age'].mean())
Datos['Embarked'] = Datos['Embarked'].fillna(Datos['Embarked'].mode()[0])
```

```python
Datos.drop('Cabin', axis=1, inplace=True)
```

```python
Datos = pd.get_dummies(Datos, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
print(Datos.isnull().sum())
```

#### 3. Seleccionar las características más relevantes para entrenar elmodelo utilizando selección de características

```python
X = Datos[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']]
y = Datos['Survived']
```

```python
print(y.value_counts())
```

#### 4. Dividir el dataset en Train y Test para evaluar correctamente el modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=0)

print(X_train.head())
print(y_train.head())
```

#### 5. Entrenar el modelo configurando los diferenteshiperparámetros.

```python
modelo_regresion = LogisticRegression(solver='liblinear')
modelo_regresion.fit(X_train, y_train)
```

```python
predicciones = modelo_regresion.predict(X_test)
```

```python
print(y_test.head())
print(predicciones[:10])
```

#### 6. Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc.

```python
print(classification_report(y_test, predicciones))
print(confusion_matrix(y_test, predicciones))
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, predicciones), annot=True, fmt='d', cmap='Blues', xticklabels=['No Sobrevivió', 'Sobrevivió'], yticklabels=['No Sobrevivió', 'Sobrevivió'])
plt.title('Matriz de Confusión')
plt.show()
```

```python
predicciones_prob = modelo_regresion.predict_proba(X_test)[:, 1]
```

```python
RocCurveDisplay.from_predictions(y_test, predicciones_prob)
plt.title('Curva ROC para la predicción de Sobrevivientes')
plt.show()
```
